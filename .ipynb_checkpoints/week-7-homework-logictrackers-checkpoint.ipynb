```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned IoT data
df = pd.read_csv("cleaned_iot_data.csv")

# Display first few rows to verify data
df.head()
```

```python
# Convert timestamp column to datetime
df["timestamp"] = pd.to_datetime(df["timestamp"])
```

```python
# Set visualization style
sns.set(style="whitegrid")
```

```python
# Limit dataset freely
df_plot = df.head(100)
```

```python
# Create Line Plot 
plt.figure(figsize=(18, 6)) # Adjust dimension to fit 100 dataset 

sns.lineplot(
    x=df_plot["timestamp"],
    y=df_plot["temperature_celsius"],
    hue=df_plot["status"],              
    marker="o"
)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add title and labels
plt.title("IoT Package Temperature Over Time", fontsize=14)
plt.xlabel("Timestamp", fontsize=12)
plt.ylabel("Temperature (°C)", fontsize=12)

# Show legend
plt.legend(title="Package Status")

# Display the plot
plt.show()
```
